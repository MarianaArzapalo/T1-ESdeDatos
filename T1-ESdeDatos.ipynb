Ej-2 Creando un DataFrame a partir de datos

```python
import pandas as pd

names = ['Bob','Jessica','Mary','John','Mel']
grades = [76,95,77,78,99]
bsdegrees = [1,1,0,0,1]
msdegrees = [2,1,0,0,0]
phddegrees = [0,1,0,0,0]

Degrees = zip(names,grades,bsdegrees,msdegrees,phddegrees)
column = ['Names','Grades','BS','MS','PhD']

df = pd.DataFrame(data = Degrees, columns=column)
df.to_csv('output/studentdates.csv', index=False, header=False)
```

Ej-4 - Creando una hoja de Excel a partir de un Dataframe

```python
import pandas as pd

names = ['Nike','Adidas','New Balance','Puma','Reebok']
grades = [176,59,47,38,99]
PriceList = zip(names,grades)

df = pd.DataFrame(data = PriceList, columns=['Names','Prices'])

df.head()
```

```python
writer = pd.ExcelWriter('output/Ejemplo4.xlsx', engine='xlsxwriter')

df.to_excel(writer, sheet_name='Sheet1')

writer.save()
```

Ej 7 - Creando una base de datos SQL a partir de un archivo CSV

```python

import pandas as pd

Location="datasets/all_050_in_06.P4.csv"

GEOID = ['111','555','555','88','454']
SUMLEV = [76,95,77,78,99]
STATE=['5','7','10','12','15']
NAME=['Hector','Maria','Jose','Carlos','Mario']
GradeList = zip(GEOID,SUMLEV, STATE, NAME)
df = pd.DataFrame(data = GradeList, columns=['GEOID', 'SUMLEV','STATE','NAME'])
df
```

```python
import os
import sqlite3 as lite

db_filename = r'output/Ej7.db'
con = lite.connect(db_filename)

df.to_sql('mytable',
    con,
    schema=None,
    if_exists='replace',
    index=True,
    index_label=None,
    chunksize=None,
    dtype=None)

con.close()
```
